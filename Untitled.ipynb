```python
def extract_words_from_pdf(pdf_file):
    words = []
    pdf_document = fitz.open(pdf_file)
    
    for page_num in range(pdf_document.page_count):
        page = pdf_document.load_page(page_num)
        text = page.get_text()
        words.extend(text.split())
        
    pdf_document.close()
    return words

```

```python
def is_english_word(word):
    return word.lower() in english_words

def check_english_words(words):
    with ThreadPoolExecutor() as executor:
        is_english = list(executor.map(is_english_word, words))
    return is_english
```

```python
def replace_punctuation(word):
    word = word.replace(',','').replace('!','').replace('.','')
```

```python
def translator_rus(word):
    translator = Translator(to_lang='ru')
    translation = translator.translate(word)
    return translation
```

```python
def translator_rus(english_word):
    translator = Translator()
    translation = translator.translate(english_word, dest='ru')
    return translation.text

def translate_words(words):
    with ThreadPoolExecutor() as executor:
        translations = list(executor.map(translator_rus, words))
    return translations
```

```python
def eng_sound(eng_word):
    language = 'en'
    tts = gTTS(text=eng_word, lang=language, slow=False)
    tts.save(f"{eng_word}.mp3")
    tts.close()
    time.sleep(100)
    return f"{eng_word}.mp3"
```
